```python
%matplotlib inline

from neuron import h
import matplotlib.pyplot as plt

h.load_file("stdrun.hoc")

soma = h.Section(name="soma")
soma.L = 20
soma.diam = 20
soma.insert("hh")

stim = h.IClamp(soma(0.5))
stim.delay = 5
stim.dur = 1
stim.amp = 0.1

t = h.Vector().record(h._ref_t)
v = h.Vector().record(soma(0.5)._ref_v)

h.finitialize(-65)
h.continuerun(40)

plt.figure()
plt.plot(t, v)
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (mV)")
plt.title("Hodgkin-Huxley soma in NEURON")
plt.show()
```
